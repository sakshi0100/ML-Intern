```
import matplotlib.pyplot as plt
import numpy as np
import pandas
df=pandas.read_csv('/content/drive/My Drive/dataset/Sakshi Singh - bitcoin.csv')
#df.count()
df['Gross profit or loss']=df['Open']-df['Close']
df
```

```
df['Date'].replace('', np.nan, inplace=True)
df.dropna(subset=['Date'], inplace=True)
df
```

```
a_values=np.array(df['Date'])
b_values=np.array(df['Gross profit or loss'])
plt.plot(a_values,b_values,'bo-',markersize=4)
plt.xlabel("Date")
plt.ylabel("Gross profit or loss")
plt.xticks(np.linspace(100,0,20),rotation=45)
plt.show()
```
